# outliers handling

```python
n_df=df.copy(deep=True)
n_df.head()
```

```python
Q1_G=n_df['Glucose'].quantile(0.25)
Q3_G=n_df['Glucose'].quantile(0.75)
IQR_G=Q3_G-Q1_G
Lower_Bound_G=Q1_G-1.5*IQR_G
Upper_Bound_G=Q3_G+1.5*IQR_G

Q1_B=n_df['BMI'].quantile(0.25)
Q3_B=n_df['BMI'].quantile(0.75)
IQR_B=Q3_B-Q1_B
Lower_Bound_B=Q1_B-1.5*IQR_B
Upper_Bound_B=Q3_B+1.5*IQR_B

Q1_A=n_df['Age'].quantile(0.25)
Q3_A=n_df['Age'].quantile(0.75)
IQR_A=Q3_A-Q1_A
Lower_Bound_A=Q1_A-1.5*IQR_A
Upper_Bound_A=Q3_A+1.5*IQR_A
```

```python
n_df.loc[df["Glucose"] < Lower_Bound_G, "Glucose"] = Lower_Bound_G
n_df.loc[df["BMI"] < Lower_Bound_B, "BMI"] = Lower_Bound_B
n_df.loc[df["BMI"] > Upper_Bound_B, "BMI"] = Upper_Bound_B
n_df.loc[df["Age"] > Upper_Bound_A, "Age"] = Upper_Bound_A
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

sns.boxplot(x=n_df['Glucose'], ax=axes[0,0]).set(title='Boxplot for Glucose')
sns.boxplot(x=n_df['BMI'], ax=axes[0,1]).set(title='Boxplot for BMI')
sns.boxplot(x=n_df['Age'], ax=axes[1,0]).set(title='Boxplot for Age')
```

# Feature engineering 

```python
x=n_df.loc[::,['Glucose','BMI','Age']]
y=n_df['Outcome']
```

```python
#Handling imbalanced dataset
from collections import Counter
print("Count of Target before over sampling: ",Counter(y))
from imblearn.over_sampling import SMOTE
os = SMOTE(sampling_strategy='minority',random_state=2)
x_over,y_over = os.fit_resample(x,y)
print("Count of Target after over sampling: ",Counter(y_over))
```

#Train & Test Split 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_over,y_over,test_size=0.3,random_state=2)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred_lr=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy Score: ",accuracy_score(y_test,y_pred_lr))
print("----------------------------------------------------------")
print("Confusion Matrix")
print(confusion_matrix(y_test,y_pred_lr))
print("----------------------------------------------------------")
print("Classification Report")
print(classification_report(y_test,y_pred_lr))
```

```python
#Visulization of Confusion Matrix - Logistic Regression
cm_lr=confusion_matrix(y_test,y_pred_lr)
group_names = ['True Neg','False Pos','False Neg','True Pos']
group_counts = ['{0:0.0f}'.format(value) for value in
                cm_lr.flatten()]
group_percentages = ['{0:.2%}'.format(value) for value in
                     cm_lr.flatten()/np.sum(cm_lr)]
labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cm_lr, annot=labels, fmt='', cmap='Blues')
```
